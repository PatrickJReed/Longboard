```python
HDB = hdbscan.HDBSCAN(min_samples=12,min_cluster_size=1000,prediction_data=True).fit_predict(umap)
```

```python
import matplotlib.pyplot as plt
clustered = (HDB >= 0)
%matplotlib inline
fig, ax = plt.subplots(1, figsize=(14, 10))
plt.scatter(umap[~clustered, 0],
            umap[~clustered, 1],
            c=(0.5, 0.5, 0.5),
            s=0.1,
            alpha=0.5)
plt.scatter(umap[clustered, 0],
            umap[clustered, 1],
            c=HDB[clustered],
            s=0.1,
            cmap='Spectral');
cbar = plt.colorbar(boundaries=np.arange(max(HDB)))
cbar.set_ticks(np.arange(max(HDB)))
```
